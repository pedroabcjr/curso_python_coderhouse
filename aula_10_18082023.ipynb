```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('fastfood.xlsx')
df.head()
```

FUNÇÕES DE AGREGAÇÃO

```python
print('count:' ,df['protein'].count()) #contagem de linhas
print('máximo:' ,df['protein'].max()) #valor maximo
print('mínimo:' ,df['protein'].min())#valor minimo
print('média:' ,df['protein'].mean())#media
print('desviopadrão:' ,df['protein'].std())#desvio padrão
print('variancia:' ,df['protein'].var())#variancia
print('soma:' ,df['protein'].sum())#soma
```

MEDIANA E QUANTILE  


```python
df.median(numeric_only=True)
```

```python
print('mediana:',df['protein'].median())
print('quantile_0:', df['protein'].quantile(0))
print('quantile_0.25:', df['protein'].quantile(0.25))
print('quantile_0.50:', df['protein'].quantile())
print('quantile_0.75:', df['protein'].quantile(0.75))
print('quantile_1:', df['protein'].quantile(1))
```

```python
df.describe().round(2)
```

GROUP BY

```python
list_macros = ['total_fat','protein', 'total_carb']
df.groupby('restaurant')[list_macros].mean().round(2)
```

```python
df.groupby('restaurant')[['calories','protein','sugar']].describe().round(2)
```

```python
df.groupby('restaurant').mean(numeric_only=True).round(2)
```

```python
df.groupby('restaurant').median(numeric_only=True).round(2)
```

```python
df.groupby('restaurant').agg({'calories':['min','max'], 'protein': ['median']})
```

```python
df.groupby('restaurant').agg({'calories': [lambda x: x.max() -x.min()], 'protein':[lambda x: x.mean().round(2) + x.median().round(2)]})
```

                                                                                 CONCATENAÇÃO

```python
acessorios_1 = ({'produto':['relogio','anel','brinco','camisa'], 'loja':['zimer','piter','piter','loyyn'], 'valor':[100,70,20,120]})
acessorios_2 = ({'produto':['gravata','colar','bolsa'], 'loja':['loyyn','piter','piter'], 'valor':[60,120,300]})

df_acess_1 = pd.DataFrame(acessorios_1)
df_acess_2 = pd.DataFrame(acessorios_2)

```

```python
df_acess_1.merge(df_acess_2,on='valor',how='left')
```

```python
pd.concat([df_acess_1,df_acess_2], ignore_index=True)
```

```python
df1 = df[df['restaurant'] == 'Mcdonalds'][['restaurant','calories', 'protein']]
df1.head()
```

```python
df2 = df[df['restaurant'] == 'Burger King'][['restaurant','calories', 'protein']]

df2.head(4)
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
pd.concat([df1.reset_index(drop=True), df2.reset_index(drop=True)], axis=1)
```
